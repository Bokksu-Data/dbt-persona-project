```python
# There's an Error present here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the query results into DataFrame
# Replace 'path_to_csv' with the actual CSV file path where you saved query results
data = pd.read_csv('Collection_table.csv')

# Preview the data
print(data.head())

# Convert dates to datetime objects for easy handling
data['first_order_date'] = pd.to_datetime(data['first_order_date'])
data['most_recent_order_date'] = pd.to_datetime(data['most_recent_order_date'])

# ===============================
# Visualization 1: Revenue by Collection
# ===============================
plt.figure(figsize=(12, 6))
sns.barplot(
    data=data.sort_values('total_revenue', ascending=False), 
    x='COLLECTION_NAMES_ARRAY', 
    y='total_revenue', 
    palette='Blues_d'
)
plt.title('Total Revenue by Collection', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Total Revenue ($)')
plt.xlabel('Collection')
plt.tight_layout()
plt.show()

# ===============================
# Visualization 2: Time-based Trends (Line Chart)
# ===============================
# Aggregate by month
time_data = data.groupby(data['first_order_date'].dt.to_period('M')).sum().reset_index()
time_data['first_order_date'] = time_data['first_order_date'].dt.to_timestamp()

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=time_data, 
    x='first_order_date', 
    y='total_revenue', 
    marker='o'
)
plt.title('Monthly Revenue Trends', fontsize=16)
plt.ylabel('Total Revenue ($)')
plt.xlabel('Month')
plt.grid(True)
plt.tight_layout()
plt.show()

# ===============================
# Visualization 3: Interactive Plot (Collections vs Revenue)
# ===============================
fig = px.bar(
    data.sort_values('total_revenue', ascending=False),
    x='COLLECTION_NAMES_ARRAY',
    y='total_revenue',
    color='brand',
    title='Revenue by Collection and Brand',
    labels={'total_revenue': 'Total Revenue ($)', 'COLLECTION_NAMES_ARRAY': 'Collection'},
    text='total_revenue'
)
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(xaxis_tickangle=-45, title_font_size=16)
fig.show()

# ===============================
# Visualization 4: Customer Engagement (Scatter Plot)
# ===============================
fig = px.scatter(
    data,
    x='unique_customers',
    y='total_revenue',
    size='total_units_sold',
    color='brand',
    hover_name='COLLECTION_NAMES_ARRAY',
    title='Customer Engagement: Revenue vs Unique Customers',
    labels={'unique_customers': 'Unique Customers', 'total_revenue': 'Total Revenue ($)'}
)
fig.update_layout(title_font_size=16)
fig.show()
```
